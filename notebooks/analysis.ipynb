```python
df = pd.read_excel('./../data/raw/northwind.xlsx', sheet_name=None)
```

```python
def to_snake_case(s):
    import re
    s = str(s).strip()
    # Replace spaces with underscores
    s = s.replace(' ', '_')
    # Convert camelCase or PascalCase to snake_case
    s = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', s)
    s = re.sub('([a-z0-9])([A-Z])', r'\1_\2', s)
    return s.lower()
    
    
def clean_columns(df):
    """
    Normalize DataFrame column names by stripping whitespace, converting to lowercase,
    and replacing spaces with underscores.

    Args:
        df (pd.DataFrame): Input DataFrame.

    Returns:
        pd.DataFrame: DataFrame with cleaned column names.
    """
    df = df.copy()
    df.columns = [to_snake_case(c) for c in df.columns]
    return df

for sheet_name, dataframe in df.items():
  df[sheet_name] = clean_columns(dataframe)

print("Column names have been cleaned for all DataFrames in the dictionary.")
```

```python
for sheet_name, dataframe in df.items():
  print(f"--- Head of {sheet_name} DataFrame ---")
  display(dataframe.head())
```

```python
for sheet_name, dataframe in df.items():
  print(f"Table: {sheet_name}")
  print("Columns:", dataframe.columns.tolist())
  print("-" * 30)
```
